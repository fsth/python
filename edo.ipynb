```python
from sympy import *
```

```python
t = Symbol('t')
y = Function('y')
phi =cos(t)*y(t)
edo =Eq(diff(y(t),t), phi )
edo
```

```python
solgen = dsolve(edo,y(t))
solgen
```

```python
t0=0
y0=1
consts = solve(Eq( y0, solgen.rhs.subs(t,t0)) , dict=True)[0]
consts
```

```python
solpar=solgen.subs(consts)
solpar
```

```python
func = lambdify(t,solpar.rhs,'numpy')
```

```python
from matplotlib.pylab import *
figure(figsize=(10,7))
xx=linspace(0,1,101)
yy=func(xx)
plot(xx,yy)
xlabel('x')
ylabel('u')
grid();
```
